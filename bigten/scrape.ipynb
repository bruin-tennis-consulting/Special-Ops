```python
BIG_URL = 'https://bigten.org/mten/schedule/?seasonFilter=2025'
# for overall table, class="MuiTable-root MuiTable-stickyHeader boost-fiakqa"
# for each row, class="MuiTableRow-root boost-1m84vbe"
```

```python
req = requests.get(BIG_URL)
soup = BeautifulSoup(req.content, "html.parser")
print(soup.prettify())
```

```python
rows = soup.find_all('tr',{"class":["MuiTableRow-root","boost-1m84vbe"]})
for r in rows:
    print(r)
```

```python
classes = []

for element in soup.find_all(class_=True):
    classes.extend(element["class"])
    
for j in classes:
    if j[0] == 'b':
        print(j)
print(classes)
```
